```python
import pandas as pd
import sqlalchemy
import matplotlib.pyplot as plt

engine = sqlalchemy.create_engine('postgresql://admin:password@db:5432/shop_db')

df = pd.read_sql("SELECT * FROM orders", engine)
```

```python
print(f"Всего заказов: {len(df)}")
category_sales = df.groupby('category')['price'].sum()
```

```python
category_sales.plot(kind='bar', title='Выручка по категориям')
plt.show()
```
